```python
import websocket
import json
from datetime import datetime, timezone

# Mobula WebSocket URL
WSS_URL = "wss://api.mobula.io"

# Your Mobula API Key
API_KEY = "e26c7e73-d918-44d9-9de3-7cbe55b63b99"

# AI Tokens you want to track
symbols = ["AI16Z", "ARC", "AVA", "SWARMS", "BUZZ", "GRIFT"]

# Keep track of printed fields
printed_fields = set()

# WebSocket Message Handling
def on_message(ws, message):
    try:
        data = json.loads(message)

        # Ensure the 'data' key exists
        if "data" not in data:
            print("‚ö† Unexpected data format:", data)
            return

        # Extract assets data
        assets = data["data"]

        # Get current time as timestamp
        current_time = datetime.now(timezone.utc).strftime('%Y-%m-%d %H:%M:%S')

        for symbol, asset in assets.items():
            # Print fields only once per token
            if symbol not in printed_fields:
                print(f"\nüìå Fields returned for {symbol}:")
                for key in asset.keys():
                    print(f"  - {key}")
                printed_fields.add(symbol)

            # Display key metrics
            name = asset.get("name", "UNKNOWN")
            price = asset.get("price", "N/A")
            market_cap = asset.get("market_cap", "N/A")
            volume_24h = asset.get("volume", "N/A")

            print(f"üìå {name} ({symbol}) | Price: ${price} | Market Cap: ${market_cap} | 24H Volume: {volume_24h} | Time: {current_time}")

    except Exception as e:
        print(f"‚ùå Error processing message: {e}")

# WebSocket Subscription
def on_open(ws):
    print("‚úÖ WebSocket Connection Opened...")

    # Print token and data structure being called
    print("\nüì° Subscribing to the following tokens:")
    for token in symbols:
        print(f"  - {token}")

    print("\nüìä Inspecting available fields per token...\n")

    # Subscription Payload
    subscribe_msg = {
        "type": "market",
        "authorization": API_KEY,
        "payload": {
            "assets": [{"symbol": symbol} for symbol in symbols],            
            "interval": 10  # Granularity of 10 seconds
        }
    }
    ws.send(json.dumps(subscribe_msg))

def on_error(ws, error):
    print(f"‚ùå WebSocket Error: {error}")

def on_close(ws, close_status_code, close_msg):
    print("üö´ WebSocket Connection Closed.")

# Start WebSocket Connection
ws = websocket.WebSocketApp(WSS_URL, on_open=on_open, on_message=on_message, on_error=on_error, on_close=on_close)
ws.run_forever()
```
